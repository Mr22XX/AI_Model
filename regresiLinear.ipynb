```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data Luas Tanah (m²) dan Harga Rumah (juta IDR)
X = np.array([30, 50, 70, 90, 110]).reshape(-1, 1)  # Variabel independen (X)
y = np.array([250, 400, 550, 700, 850])  # Variabel dependen (y)

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, y)  # Melatih model dengan data

# Prediksi harga rumah untuk luas tanah tertentu
luas_baru = np.array([[60]])  # Contoh luas tanah 60 m²
prediksi_harga = model.predict(luas_baru)

print(f"Prediksi harga rumah untuk {luas_baru[0][0]} m²: {prediksi_harga[0]:.2f} juta IDR")

# Visualisasi hasil regresi linear
plt.scatter(X, y, color='blue', label="Data Asli")  # Plot data asli
plt.plot(X, model.predict(X), color='red', label="Garis Regresi")  # Plot garis regresi
plt.scatter(luas_baru, prediksi_harga, color='green', marker='o', label="Prediksi 60 m²")

plt.xlabel("Luas Tanah (m²)")
plt.ylabel("Harga Rumah (juta IDR)")
plt.title("Regresi Linear: Prediksi Harga Rumah")
plt.legend()
plt.show()
```
